```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
```

```python
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
```

```python
cosy = np.cos(x)
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny, label='sin(x)')
plt.plot(x, cosy, color='red', linestyle='--', label='cos(x)')
plt.plot(x, np.zeros(100),color='black')
plt.xlabel('x axis')
plt.ylabel('y value')
plt.legend()
plt.title('sin/cos function')
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy)
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)

plt.scatter(x, y)
plt.axis([-5,5,-5,5])
```
