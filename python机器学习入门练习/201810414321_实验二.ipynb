### 1.递归与递推

问题1：猴子爬山一只顽猴在一座有50级台阶的小山上爬山跳跃。上山时需从山脚至山顶往上跳50级台阶，一步可跳1级，或跳2级，或跳3级，求上山有多少种不同的跳法？ （请用递归和递推方法分别实现，并比较两者的运行效率）



```python
import time


# -*- coding: utf-8 -*-
# 递归
def func_1(x):
    if x == 1:
        return 1
    elif x == 2:
        return func_1(1) + 1
    elif x == 3:
        return func_1(2) + func_1(1) + 1
    else:
        return func_1(x-1) + func_1(x-2) + func_1(x-3)
    
# 递推
def func_2(x):
    result = [0]
    result.append(1)
    result.append(2)
    result.append(4)
    for i in range(4, x + 1):
        result.append(result[i-1] + result[i-2] + result[i-3])
    return result[-1]

start = time.time()
step = func_1(36)
end = time.time()
print("36层台阶递归结果：", step, "耗时：", end - start)

start = time.time()
step = func_2(100)
end = time.time()
print("100层台阶递推结果：", step, "耗时：", end - start)
        
```

### 2.贪心算法

问题2：贪心算法：假设玩具小汽车x元一辆，小明用y元钱买z辆，
请问如何找零使得找零的纸币的张数最少！

```python
#code

def charge(n): # n表示需要找零的钱数
    money = [50, 20, 10, 5, 1]# 面值
    num = 0# 找零数量
    print("需要找零:")
    for i in range(5):
        num = n // money[i]
        n = n % money[i]
        print(money[i], "元", num, "张")
    
try:
    # 小汽车单价
    x = int(input("请输入小汽车单价："))
    # 预算
    y = int(input("请输入预算："))
    # 需要买的数量
    z = int(input("请输入打算买的数量"))
    
    # 需要找零的金额
    n = y - x * z
    
    if n < 0:
        print("您的预算不足！")
    else:
        charge(n)
        
except:
        print("无效数据，请重新输入！")
```

### 3.动态规划

问题3：一个小偷打劫一个保险箱，保险箱中有5中物品，每种物品的重量不同价值也不同，物品的重量与价值如表5.2所示，小偷的背包只能负重8公斤，要怎样才能使偷走的物品总价值最大？

```python
# bag表示物品的重量与其所对应的价值，n为背包最大容量
bag = [[4,45],[5,57],[2,22],[1,11],[6,67]]
n = 8
dp = []

for i in range(len(bag)): dp.append([0]*(n+1))

for i in range(n+1):

    if i >= bag[0][0]: dp[0][i] = bag[0][1]

    #第一次遍历数组将得到第一个物品所对应的最大质量得出

for i in range(1,len(bag)):

    for j in range(n+1):

        if bag[i][0] <= j:

            dp[i][j] = max(dp[i-1][j], dp[i-1][j-bag[i][0]]+bag[i][1])

            # 取dp[i][j]当前的最大质量
j = n
for i in range(len(bag)-1, -1, -1):
    if dp[i][j] > dp[i-1][j]:
        print("携带物品", i+1, "  重量：", bag[i][0], "价值：", bag[i][1])
        j = j - bag[i][0]
print("最大收益", dp[-1][-1]) # 打印dp最终j=n(背包最大容量)的最大质量   
```
