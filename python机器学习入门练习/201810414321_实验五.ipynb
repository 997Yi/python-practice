## 实验步骤

### 1鸢尾花数据分类

```python
#导入第三方库
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
```

#### 1.1 数据集读取

```python
#code
iris = load_iris()
print(iris)
X = iris.data[:, :2]
Y = iris.target
print(X,Y)
```

#### 1.2数据预处理

```python
#code
h = 0.02
x_min,x_max = X[:,0].min()-0.5,X[:,0].max()+0.5
y_min,y_max = X[:,1].min()-0.5,X[:,1].max()+0.5
xx,yy = np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))
```

#### 1.3模型构建

```python
#code
lr = LogisticRegression(C=1e5)#正则化参数倒数
```

#### 1.4模型训练

```python
lr.fit(X, Y)
```

#### 1.6.模型可视化

```python
Z = lr.predict(np.c_[xx.ravel(),yy.ravel()])
Z = Z.reshape(xx.shape)
# 绘制
plt.figure(1,figsize=(8,6))
plt.pcolormesh(xx,yy,Z,cmap=plt.cm.Paired)
plt.scatter(X[:50,0],X[:50,1],color='red',marker='o',label='setosa')
plt.scatter(X[50:100,0],X[50:100,1],color='blue',marker='x',label='versicolor')
plt.scatter(X[100:,0],X[100:,1],color='green',marker='s',label='Virginica')
# 坐标轴名称
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
# 坐标轴刻度
plt.xlim(xx.min(),xx.max())
plt.ylim(yy.min(),yy.max())
plt.xticks()
plt.yticks()
# 图例
plt.legend(loc=2)
plt.show()
```

### 2.手写数字识别

### KNN，逻辑回归，贝叶斯分类对比

### 2.1导入第三方库

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn import naive_bayes, datasets
from sklearn import metrics
from sklearn import cross_decomposition
```

### 2.2数据导入及预处理

#### 数据导入

```python
digits = datasets.load_digits()
# print(digits)
data = digits.data
target = digits.target
images = digits.images
```

#### 数据集划分

```python
# print(len(data))
X_train = data[:1200]
Y_train = target[:1200]
```

#### 测试机

```python
x_test = data[1200:]
y_true = target[1200:]
```

### 2.3建立分类模型

#### 逻辑回归模型

```python
lrclf = LogisticRegression(solver='lbfgs')
```

#### KNN模型

```python
knnclf = KNeighborsClassifier(n_neighbors=5)
```

#### 贝叶斯模型 

```python
NBclf = naive_bayes.GaussianNB()
```

### 2.4模型训练

```python
lrclf.fit(X_train,Y_train)
knnclf.fit(X_train,Y_train)
NBclf.fit(X_train,Y_train)
```

### 2.5模型测试

```python
y_pre_lr = lrclf.predict(x_test)
y_pre_knn = knnclf.predict(x_test)
y_pre_NB = NBclf.predict(x_test)
```

### 2.6算法性能评估比对

#### 第一种获取分类准确率

```python
print("lr_score=",lrclf.score(x_test,y_true))
print("knn_score=",knnclf.score(x_test,y_true))
print("NB_score=",NBclf.score(x_test,y_true))
```

#### 第二种获取分类准确率方法

```python
print("逻辑回归分类正确率=",metrics.accuracy_score(y_true,y_pre_lr))
# 只能用于二分类，多分类如何解决？是否有必要解决
# print("逻辑回归roc_auc=",metrics.roc_auc_score(y_true,y_pre_lr))
# print("逻辑回归F1_score=",metrics.f1_score(y_true,y_pre_lr))
```

### 2.7可视化分类结果示例

```python
plt.figure(figsize=(32,32)) # 新建画布大小
for i in range(100):
    plt.subplot(10,10,i+1)  # 将画布分成10*10显示区域做100个子图
    plt.imshow(x_test[i].reshape(8,8))    # 每张图片8*8像素
    plt.axis('off')     # 隐藏坐标轴
    title = 'lr:'+str(y_pre_lr[i])+'\nKNN'+str(y_pre_knn[i])+'\nNB'+str(y_pre_NB[i])+'\nTrue'+str(y_true[i])
    plt.title(title)
```
